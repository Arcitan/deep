## Setup code

Here we import the libraries we need and add some setup code to make our plots look a bit nicer.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.dpi"] = 150
plt.rcParams["figure.figsize"] = 10, 6
```

## Line Plot

```python
f = lambda x: 3 * x**2 * np.cos(x) + np.exp(-x)
```

```python
import warnings
```

```python
x = np.arange(100)
y = f(x)

# use this context manager to make "xkcd-style" plots!
with plt.xkcd():
    # we might be missing some xkcd fonts, so we'll ignore those
    with warnings.catch_warnings():
        warnings.simplefilter("ignore") 
        plt.plot(x, (y / y.max() + 1) * 2)
    plt.title("Expected GPA vs. Semester Progress")
    plt.ylabel("E[GPA]")
    plt.xlabel("Days from Semester Start")
    plt.show()
```
